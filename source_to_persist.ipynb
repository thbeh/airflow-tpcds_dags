```python
import pandas as pd
import pyarrow as pa
import pyarrow.parquet as pq

from io import StringIO


from s3fs import S3FileSystem

s3 = S3FileSystem(key='minio', secret='minio123', client_kwargs={'endpoint_url':'http://minio:9000'})
```

```python
pd.set_option('display.max_columns', None)
source_bucket = 'source-data'
target_bucket = 'persist-data'
```

```python
def write_pandas_parquet_to_s3(df, bucket, key, file):
    # dummy dataframe
    table = pa.Table.from_pandas(df)
    pq.write_to_dataset(table, 's3://{0}/{1}/event={2}'.format(bucket,key,file), filesystem=s3, compression='snappy')
```
